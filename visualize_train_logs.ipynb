```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pathlib

import utils
```

```python
model = 'resnet'
out_file = pathlib.Path.cwd() / 'results' / model / 'train_logs.txt'
tiny_log = utils.readProgressLog(out_file)
epoch_arr = np.arange(len(tiny_log['train']))
```

```python
# xrange = 800
xrange = len(epoch_arr)
colorList = ['b', 'r', 'orange']
labelList = [r'Training', r'Test set', r'Clean subset']

fig = plt.figure(figsize=(8,6))
ax = plt.subplot(111)  
for i, setkey in enumerate(tiny_log.keys()):
    ax.plot(epoch_arr[:xrange], tiny_log[setkey][:xrange], lw = 2, color = colorList[i], label=labelList[i])

ax.legend(loc=4, bbox_to_anchor=(0.98, 0.04), fontsize='large',
          frameon=False, fancybox=True, shadow=True, ncol=1)
ax.set_ylabel('% Accuracy')
ax.set_xlabel('epoch')

plt.show()
```
